# Computing the Matrix profile of two time series using Stumpy 
This example demonstrates how to use the Matrix Profile to measure the similarity between two time series. The resulting plot allows you to visually inspect how similar the two time series are at different points.

```python
import numpy as np
import stumpy
import matplotlib.pyplot as plt

# Generate two synthetic time series
np.random.seed(42)
time_series_1 = np.sin(np.linspace(0, 50, 100)) + np.random.normal(0, 0.1, 100)
n = np.random.randint(0,100)
time_series_2 = np.sin(np.linspace(0, 30, 100)) + np.random.normal(0, 0.1, 100) + n  # Slightly different

# Define the subsequence length (window size)
m = 20

# Compute the Matrix Profile between the two time series
matrix_profile = stumpy.stump(T_A=time_series_1, m=m, T_B=time_series_2)

# The matrix_profile is a 2D array where the first column is the Matrix Profile itself
# and the second column is the index of the nearest neighbor subsequence in the other time series
profile, indices = matrix_profile[:, 0], matrix_profile[:, 1]


```

```python
# Plot the time series and the Matrix Profile
plt.figure(figsize=(12, 6))

plt.subplot(3, 1, 1)
plt.plot(time_series_1, label='Time Series 1')
plt.legend()

plt.subplot(3, 1, 2)
plt.plot(time_series_2, label='Time Series 2')
plt.legend()

plt.subplot(3, 1, 3)
plt.plot(profile, label='Matrix Profile')
plt.xlabel('Subsequence Start Index')
plt.ylabel('Distance')
plt.legend()

plt.tight_layout()
plt.show()
```
