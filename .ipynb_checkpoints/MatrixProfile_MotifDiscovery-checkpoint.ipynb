# Matrix profile - Motif discovery 

- Let's explore how to perform motif discovery using the Matrix Profile on a real-world dataset. For this example, we'll use stock market data, specifically the closing prices of Apple Inc. (AAPL) over the past few years. We'll identify recurring patterns (motifs) in the stock price time series.

```python
import numpy as np
import pandas as pd
import yfinance as yf
import stumpy
import matplotlib.pyplot as plt
from datetime import datetime, timedelta
```

```python
# 1. Data Collection: Fetch AAPL Stock Data
end_date = datetime.today()
start_date = end_date - timedelta(days=365 * 2)  # Fetch data for the past 2 years

# Download the data
data = yf.download('AAPL', start=start_date.strftime('%Y-%m-%d'), end=end_date.strftime('%Y-%m-%d'))

# Use the closing price as the time series
time_series = data['Close'].values

# 2. Preprocessing: Handle Missing Data
# Check for NaN values and handle them (if any)
if np.isnan(time_series).any():
    time_series = pd.Series(time_series).fillna(method='ffill').fillna(method='bfill').values

# 3. Define the Subsequence Length (Window Size)
# Choosing an appropriate window size is crucial.
# For stock data, a window of 30 days (~1 month) can be a good starting point.
m = 30

# 4. Compute the Matrix Profile
mp = stumpy.stump(time_series, m)

# The first column of mp contains the Matrix Profile values
matrix_profile = mp[:, 0]

# The second column contains the indices of the nearest neighbor for each subsequence
profile_indices = mp[:, 1]

# 5. Motif Discovery: Identify the Top Motifs
# Find the indices of the lowest Matrix Profile values (i.e., most similar subsequences)
num_motifs = 2  # Number of motifs to find
motif_indices = np.argsort(matrix_profile)[:num_motifs]

# Extract the motifs and their matching subsequences
motifs = []
for idx in motif_indices:
    motif = time_series[idx:idx + m]
    nearest_neighbor_idx = profile_indices[idx]
    nearest_neighbor = time_series[nearest_neighbor_idx:nearest_neighbor_idx + m]
    motifs.append((idx, motif, nearest_neighbor_idx, nearest_neighbor))

# 6. Visualization
plt.figure(figsize=(15, 10))

# Plot the entire time series
plt.subplot(len(motifs) + 1, 1, 1)
plt.plot(time_series, label='AAPL Closing Prices')
plt.title('AAPL Closing Prices Over Time')
plt.xlabel('Time (Days)')
plt.ylabel('Price (USD)')
plt.legend()

# Plot each motif and its nearest neighbor
for i, (idx, motif, nn_idx, nearest_neighbor) in enumerate(motifs):
    plt.subplot(len(motifs) + 1, 1, i + 2)
    plt.plot(range(idx, idx + m), motif, label=f'Motif {i + 1}')
    plt.plot(range(nn_idx, nn_idx + m), nearest_neighbor, label=f'Nearest Neighbor to Motif {i + 1}')
    plt.title(f'Motif {i + 1} and Its Nearest Neighbor')
    plt.xlabel('Time (Days)')
    plt.ylabel('Price (USD)')
    plt.legend()

plt.tight_layout()
plt.show()

# 7. Print Out Motif Details
for i, (idx, _, nn_idx, _) in enumerate(motifs):
    motif_start_date = data.index[idx].strftime('%Y-%m-%d')
    motif_end_date = data.index[idx + m - 1].strftime('%Y-%m-%d')
    nn_start_date = data.index[nn_idx].strftime('%Y-%m-%d')
    nn_end_date = data.index[nn_idx + m - 1].strftime('%Y-%m-%d')
    print(f"Motif {i + 1}: {motif_start_date} to {motif_end_date}")
    print(f"Nearest Neighbor {i + 1}: {nn_start_date} to {nn_end_date}\n")
```
